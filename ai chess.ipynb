```
#define is diagonal - returns true or false stating to_state is diagonal from from_state
#takes chessboard as input, from_state, to_state
def is_diagonal(chessboard, starting, location):
  
  
  left_D_up = starting
  while True:
    #make new location left 
    left_D_up = ( (left_D_up[0]-1), (left_D_up[1]-1) )

    if left_D_up == (location):
      return True
    elif 8 > left_D_up[0] > 0 and 8 >left_D_up[1] > 0:
      continue
    else:
      break


  right_D_up = starting
  while True:
    #make new location left 
    right_D_up = ( (right_D_up[0]-1), (right_D_up[1]+1) )

    if right_D_up == (location):
      return True
    elif 8 > right_D_up[0] > 0 and 8 >right_D_up[1] > 0:
      continue
    else:
      break


  #starting point to go left
  left_D_down = starting
  while True:
    #make new location left 
    left_D_down = ( (left_D_down[0]+1), (left_D_down[1]-1) )

    if left_D_down == (location):
      return True
    elif 8 > left_D_down[0] > 0 and 8 > left_D_down[1] > 0:
      continue
    else:
      break

  right_D_down = starting

  while True:

    right_D_down = ( (right_D_down[0]+1), (right_D_down[1]+1) )

    if right_D_down == (location):
      return True
    elif 8 > right_D_down[0] > 0 and 8 > right_D_down[1] > 0:
      continue
    else:
      break

  return False
```
